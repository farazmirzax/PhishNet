# Step 1: Data Loading & Initial Health Check

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define the path dynamically
# This looks for 'phishing.csv' inside 'backend/data/raw/'
base_dir = os.path.dirname(os.getcwd())
file_path = os.path.join(base_dir, '..', 'data', 'raw', 'phishing.csv')

print(f"📂 Looking for data at: {file_path}")

# 2. Load the Data
try:
    df = pd.read_csv(file_path)
    print("✅ Data Loaded Successfully!")
    print(f"📊 Total Rows: {len(df)}")
    
    # 3. Peek at the data (First 5 rows)
    print("\n--- FIRST 5 ROWS ---")
    display(df.head())

    # 4. Check for Class Balance (Phishing vs Legitimate)
    # We need to know the name of the target column first, 
    # so we grab the very last column automatically.
    target_col = df.columns[-1] 
    print(f"\n🎯 Target Column identified as: '{target_col}'")

    plt.figure(figsize=(6,4))
    sns.countplot(x=df[target_col])
    plt.title(f"Class Balance: {target_col}")
    plt.show()

    # 5. Check for Missing Values
    print("\n--- MISSING VALUES ---")
    print(df.isnull().sum().sum())

except FileNotFoundError:
    print("❌ Error: File not found. Please check that 'phishing.csv' is in 'backend/data/raw/'")
```

# Step 2: Clean, Select, and Split 🧹

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# 1. Define the Robust Features we want to keep
# These are easy to calculate in Python for our API later
NUMERIC_FEATURES = [
    'length_url', 'length_hostname', 'ip', 'nb_dots', 'nb_hyphens', 
    'nb_at', 'nb_qm', 'nb_and', 'nb_eq', 'nb_slash', 'nb_colon', 'ratio_digits_url'
]

# 2. Select the Data
# X_url = The raw text (for LSTM)
# X_meta = The derived numbers (for Dense)
# y = The label
data_clean = df[['url'] + NUMERIC_FEATURES + ['status']].copy()

# 3. Encode the Target (Legitimate=0, Phishing=1)
# NOTE: We force 'phishing' to be 1 because that's what we want to detect
data_clean['target'] = data_clean['status'].apply(lambda x: 1 if x == 'phishing' else 0)

# 4. Split the Data (80% Train, 20% Test)
# we use random_state=42 so we get the same split every time
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    data_clean.drop(['status', 'target'], axis=1),
    data_clean['target'],
    test_size=0.2,
    random_state=42,
    stratify=data_clean['target'] # Keep the 50/50 balance in both sets
)

print(f"✅ Data Split Complete!")
print(f"Train Shape: {X_train_raw.shape}")
print(f"Test Shape:  {X_test_raw.shape}")
print(f"Features selected: {len(NUMERIC_FEATURES)}")

# Verify the class split
print("\nTrain Class Balance:")
print(y_train.value_counts())
```

# Step 3: The "Translator" (Preprocessing) 🗣️

```python
import numpy as np
import pickle
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import StandardScaler
import joblib

# --- CONSTANTS ---
MAX_VOCAB_SIZE = 5000  # Only learn the top 5,000 most common URL "words"
MAX_SEQUENCE_LENGTH = 150 # Cut off URLs longer than this (or pad shorter ones)

print("🔄 Starting Preprocessing...")

# ---------------------------------------------------------
# PART A: PROCESS THE URLS (Text -> Numbers)
# ---------------------------------------------------------
# 1. Initialize the Tokenizer (Character-level or Word-level?)
# For URLs, 'char_level=True' is usually better to catch typos (g0ogle vs google)
tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, char_level=True, lower=True)

# 2. Fit on the Training Data ONLY (Never fit on test data!)
tokenizer.fit_on_texts(X_train_raw['url'])

# 3. Convert text to sequences of numbers
X_train_seq = tokenizer.texts_to_sequences(X_train_raw['url'])
X_test_seq = tokenizer.texts_to_sequences(X_test_raw['url'])

# 4. Pad sequences so they are all the same length (150)
X_train_pad = pad_sequences(X_train_seq, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

print(f"✅ Text Processed. Shape: {X_train_pad.shape}")
print(f"   Example: 'google.com' might look like: {X_train_pad[0][:10]}...")


# ---------------------------------------------------------
# PART B: PROCESS THE METADATA (Numbers -> Scaled Numbers)
# ---------------------------------------------------------
# 1. Initialize Scaler
scaler = StandardScaler()

# 2. Select just the numerical columns
numeric_cols = NUMERIC_FEATURES # from previous cell

# 3. Fit on Train, Transform both
X_train_meta = scaler.fit_transform(X_train_raw[numeric_cols])
X_test_meta = scaler.transform(X_test_raw[numeric_cols])

print(f"✅ Metadata Scaled. Shape: {X_train_meta.shape}")


# ---------------------------------------------------------
# PART C: SAVE THE ARTIFACTS (CRITICAL!)
# ---------------------------------------------------------
# We MUST save the Tokenizer and Scaler. 
# The API needs them to process new user links later.
processors_dir = os.path.join(base_dir, '..', 'data', 'processed')
os.makedirs(processors_dir, exist_ok=True)

# Save Tokenizer
with open(os.path.join(processors_dir, 'tokenizer.pickle'), 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)

# Save Scaler
joblib.dump(scaler, os.path.join(processors_dir, 'scaler.joblib'))

print(f"💾 Processors saved to: {processors_dir}")
```

# Step 4: Building the "Y-Network" 🧠

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Concatenate, Dropout, Bidirectional
from tensorflow.keras.utils import plot_model

# --- ARCHITECTURE DESIGN ---

# 1. INPUT A: The URL Sequence (Text)
# Shape: (150,) because we padded everything to length 150
input_seq = Input(shape=(MAX_SEQUENCE_LENGTH,), name='url_input')

# Embedding Layer: Turns integers [4, 12, 99] into rich vectors
# Input dim = 5000 (vocab size), Output dim = 32 (vector size)
x1 = Embedding(input_dim=MAX_VOCAB_SIZE, output_dim=32, name='embedding')(input_seq)

# Bi-Directional LSTM: Reads URL from Left->Right AND Right->Left
x1 = Bidirectional(LSTM(32, return_sequences=False, name='lstm_layer'))(x1)


# 2. INPUT B: The Metadata (Numbers)
# Shape: (12,) because we selected 12 numeric features
input_meta = Input(shape=(len(NUMERIC_FEATURES),), name='meta_input')

# Dense Layers for the numbers
x2 = Dense(64, activation='relu')(input_meta)
x2 = Dense(32, activation='relu')(x2)


# 3. MERGE (The Y-Shape)
# We glue the LSTM output (64 units) and Dense output (32 units) together
merged = Concatenate(name='concat_layer')([x1, x2])

# 4. THE COMMON BRAIN (Deep Reasoning)
x3 = Dense(64, activation='relu')(merged)
x3 = Dropout(0.5)(x3) # Prevents overfitting (randomly shuts off neurons)


# 5. OUTPUT
# Sigmoid: Returns a value between 0 (Legit) and 1 (Phishing)
output = Dense(1, activation='sigmoid', name='output')(x3)


# 6. BUILD THE MODEL
model = Model(inputs=[input_seq, input_meta], outputs=output)

# 7. COMPILE
model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

print("✅ Model Architecture Built!")
model.summary()
```

# Step 5: The Training Loop (The Moment of Truth) 🔥

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# --- TRAINING CONFIGURATION ---

# 1. Define Callbacks (The Referees)
# EarlyStopping: Stop if validation loss doesn't improve for 3 epochs
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Checkpoint: Save the BEST version of the model (not just the last one)
models_dir = os.path.join(base_dir, '..', 'models')
os.makedirs(models_dir, exist_ok=True)
checkpoint_path = os.path.join(models_dir, 'phishnet_v1.keras')

checkpoint = ModelCheckpoint(checkpoint_path, 
                             monitor='val_accuracy', 
                             save_best_only=True,
                             verbose=1)

print(f"🚀 Starting Training... Models will be saved to: {models_dir}")

# 2. TRAIN THE MODEL
# We feed it TWO inputs: {'url_input': X_train_pad, 'meta_input': X_train_meta}
history = model.fit(
    x={'url_input': X_train_pad, 'meta_input': X_train_meta},
    y=y_train,
    validation_data=(
        {'url_input': X_test_pad, 'meta_input': X_test_meta}, 
        y_test
    ),
    epochs=20,          # Maximum loops
    batch_size=32,      # Learn from 32 links at a time
    callbacks=[early_stop, checkpoint],
    verbose=1
)

print("✅ Training Complete!")
```
